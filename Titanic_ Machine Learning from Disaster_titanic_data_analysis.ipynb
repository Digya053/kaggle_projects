**Handling missing values**

```
sns.heatmap(df.isnull(), cbar = False).set_title("Missing values heatmap")
```

Thus, four of the attributes (Age, Fare, Cabin and Embarked) contain null values.

```
ax = df_train["Age"].hist(bins=70, color='blue', alpha=0.5)
ax.set(xlabel='Age', ylabel='Count')
plt.show()
```

We can see that the plot of age is right skewed. So, imputing the missing values with mean can give biased result. Thus, we will impute it with median value.

```
df["Age"] = df["Age"].fillna(df["Age"].median(skipna=True))
```

```
df['Cabin'].fillna("N", inplace=True)
```

The cabin names start with an English alphabet. So, we will replace the cabin names by N and extract only the initial alphabet.

```
df['Cabin'] = [i[0] for i in df['Cabin']]
```

```
df['Cabin'].value_counts()
```

We will replace the missing value in 'Embarked' with most reoccuring value.

```
df['Embarked'].value_counts()
```

```
df['Embarked'] = df['Embarked'].fillna('S')
```

For filling the missing values in 'Fare', we will compute the average 'Fare' of the rows with the similar values as the row with null value.

```
df[df["Fare"].isnull()]
```

```
df.head()
```

```
df.query('Pclass == 3 and Sex == "male" and Age > 60 and Parch == 0 and SibSp == 0 and Cabin == "N" and Embarked == "S"')
```

```
avg_fare = (df.loc[326, 'Fare'] + df.loc[851, 'Fare'])/2
avg_fare
```

```
df['Fare'] = df['Fare'].fillna(avg_fare)
```

We don't have any null values now.

Now, we will add new columns and modify the values in existing columns for making those features more useful.

```
def create_relatives_series(df):
  df['relatives'] = df['SibSp'] + df['Parch']
  return df
df_train = create_relatives_series(df_train)
df = create_relatives_series(df)
```

```
data = [df, df_train]
titles = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}

for dataset in data:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr',\
                                            'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    dataset['Title'] = dataset['Title'].map(titles)
    dataset['Title'] = dataset['Title'].fillna(0)
```

```
data = [df, df_train]

for dataset in data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[(dataset['Fare'] > 31) & (dataset['Fare'] <= 99), 'Fare']   = 3
    dataset.loc[(dataset['Fare'] > 99) & (dataset['Fare'] <= 250), 'Fare']   = 4
    dataset.loc[ dataset['Fare'] > 250, 'Fare'] = 5
    dataset['Fare'] = dataset['Fare'].astype(int)
```
